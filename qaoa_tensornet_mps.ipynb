```python
#Note to self:
#...Ideally, for a fixed number of nodes, we want to average over several graphs (labelled by random seed)

%matplotlib inline

import cudaq
from cudaq import spin

from typing import List
import numpy as np
import time
import igraph as ig
from igraph import Graph
import matplotlib.pyplot as plt
import random
import sys

import os
#os.environ['PYDEVD_DISABLE_FILE_VALIDATION'] = '1'
#os.environ['CUDA_LAUNCH_BLOCKING'] = "1"
#os.environ['CUDAQ_MPS_MAX_BOND'] = '16' #default is 64

#print(cudaq.get_target()) #show all available backends
cudaq.set_target('qpp-cpu')
#cudaq.set_target('nvidia', option = 'mgpu')
#cudaq.set_target('tensornet-mps')

# Here we build up a kernel for QAOA with `p` layers, with each layer
# containing the alternating set of unitaries corresponding to the problem
# and the mixer Hamiltonians. The algorithm leverages the VQE algorithm
# to compute the Max-Cut of a rectangular graph illustrated below.

#       v0  0---------------------0 v1
#           |                     |
#           |                     |
#           |                     |
#           |                     |
#       v3  0---------------------0 v2
# The Max-Cut for this problem is 0101 or 1010.

# Problem parameters.
max_node_count: int = 6
#node_count: int = 4
#qubit_count: int = node_count
layer_count: int = 7
parameter_count: int = 2 * layer_count
graph_reps: int = 5
colours: list = ["blue", "red"]

# The problem Hamiltonian
#hamiltonian = spin.z(0) * spin.z(1) + spin.z(1) * spin.z(2) + spin.z(0) * spin.z(3) + spin.z(2) * spin.z(3)
#print(type(hamiltonian))

@cudaq.kernel
def kernel_qaoa(qubit_count: int, layer_count: int, thetas: List[float]):
    """QAOA ansatz for Max-Cut"""
    qvector = cudaq.qvector(qubit_count)

    # Create superposition
    h(qvector)

    # Loop over the layers
    for layer in range(layer_count):
        # Loop over the qubits
        # Problem unitary
        for qubit in range(qubit_count):
            x.ctrl(qvector[qubit], qvector[(qubit + 1) % qubit_count])
            rz(2.0 * thetas[layer], qvector[(qubit + 1) % qubit_count])
            x.ctrl(qvector[qubit], qvector[(qubit + 1) % qubit_count])

        # Mixer unitary
        for qubit in range(qubit_count):
            rx(2.0 * thetas[layer + layer_count], qvector[qubit])

node_list: list = []
max_times: list = []
min_times: list = []
avg_times: list = []

for num_nodes in range(6, max_node_count+1):

    qubit_count: int = num_nodes
    node_list.append(num_nodes)
    tms: list = []

    for cur_iter in range(graph_reps):
        #Set seeds to make each problem instance repeatable. Note in the list of parameters, we first enumerate the all angles for the problem unitary before the ones for the mixer unitary
        cudaq.set_random_seed(cur_iter) #random seed apparently not available for tensornet-mps backend - double check.
        np.random.seed(cur_iter)
        random.seed(cur_iter)

        #cur_graph = ig.Graph.Erdos_Renyi(n = num_nodes,  p = 0.7, directed = False, loops = False) #generates graph where every edge has independent probability p of existing
        #cur_graph = ig.Graph.Erdos_Renyi(n = num_nodes, m = node_count / 2, directed=False, loops=False) #generates graph where m edges are uniformly chosen at random
        cur_graph = Graph.K_Regular(num_nodes, 3, directed = False, multiple = False) #generates k-regular graph. NB either k or number of nodes must be even.

        #ig.summary(cur_graph)
        #sys.exit(0)

        hamiltonian: cudaq.SpinOperator = 0

        for edge in cur_graph.es:
            #print(f"Edge from {edge.source} to {edge.target} (ID: {edge.index})")
            hamiltonian += 0.5 * (spin.z(edge.source) * spin.z(edge.target) - 1)

        print(hamiltonian)

        optimizer = cudaq.optimizers.COBYLA()
        optimizer.initial_parameters = np.random.uniform(-np.pi / 8.0, np.pi / 8.0, parameter_count)
        #optimizer.initial_parameters = np.array([1.0]*parameter_count)
        print(f"Initial parameters = {optimizer.initial_parameters}\n")

        # Define the objective, return `<state(params) | H | state(params)>`. Note the hamiltonian is already diagonal in the computational basis so the expectation returns an eigenvalue.
        def objective(parameters):
            return cudaq.observe(kernel_qaoa, hamiltonian, qubit_count, layer_count,
                                parameters).expectation()

        # Optimize!
        start_time = time.time()

        optimal_expectation, optimal_parameters = optimizer.optimize(
            dimensions=parameter_count, function=objective)

        run_time = time.time() - start_time
        tms.append(run_time)
        print(f"Run took {run_time}s")

        # Print the optimized value and its parameters
        print(f"Max cut value = {-1 * optimal_expectation}")
        print(f"Optimal parameters = {optimal_parameters}")

        # Sample the circuit using the optimized parameters
        counts = cudaq.sample(kernel_qaoa, qubit_count, layer_count, optimal_parameters)
        print(counts)

        optimal_bitstring = counts.most_probable()
        print(f"Optimal bitstring is {optimal_bitstring}")

        for i in range(num_nodes):
            cur_graph.vs[i]["color"] = colours[int(optimal_bitstring[i])]

        fig, ax = plt.subplots()
        ig.plot(cur_graph, target=ax, layout="circle", vertex_size=8)
        plt.show()

    max_times.append(max(tms))
    min_times.append(min(tms))
    avg_times.append(np.mean(np.array(tms)))

plt.xticks(range(int(min(node_list)), int(max(node_list)+1)))
plt.xlabel('Number of nodes')
plt.ylabel('Run time (s)')
plt.plot(node_list, max_times, '.', label = 'Max time')
plt.plot(node_list, min_times, '.', label = 'Min time')
plt.plot(node_list, avg_times, '.', label = 'Average time')
plt.show()
```
